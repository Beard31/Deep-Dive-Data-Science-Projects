```python
# Step 1: Project to a CRS that uses meters (Web Mercator)
schools_proj = all_schools.to_crs(epsg=4326)

# Step 2: Compute centroids in projected CRS
schools_proj['centroid'] = schools_proj.geometry.centroid

# Step 3: Convert centroids back to WGS84 for folium
schools_proj = schools_proj.set_geometry('centroid')
schools_wgs = schools_proj.to_crs(epsg=4326)

# Step 4: Merge centroids back into original polygon GeoDataFrame (also in EPSG:4326)
all_schools_wgs = all_schools.to_crs(epsg=4326)
all_schools_wgs['centroid'] = schools_wgs.geometry
```

```python
import folium
for _, row in all_schools_wgs.iterrows():
    point = row["centroid"]
    if point and point.geom_type == 'Point':
        folium.Marker(
            location=[point.y, point.x],
            popup=f"{row['SCHOOLSERV']} ({row['GRADELEVEL']})",
            icon=folium.Icon(color=(
                "blue" if row["GRADELEVEL"] == "high" else
                "green" if row["GRADELEVEL"] == "middle" else
                "orange"))
        ).add_to(marker_cluster)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Create GeoDataFrame from Plotly data
gdf_incidents = gpd.GeoDataFrame(
    incidents_500,
    geometry=gpd.points_from_xy(incidents_500["longitude"], incidents_500["latitude"]),
    crs="EPSG:4326"
)

# Add incident markers to folium
for _, row in gdf_incidents.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=5,
        color="red",
        fill=True,
        fill_opacity=0.6,
        popup=row["type"]
    ).add_to(marker_cluster)
```
